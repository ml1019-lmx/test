```python
def a_move(lst):
    lst1=lst[:]
    lst1[1]= str(float(lst1[1])+8.76000000000)
    return lst1
    
```

```python
def b_move(lst):
    lst2=lst[:]
    lst2[1]= str(float(lst2[1])+4.3800000000)
    lst2[2]= str(float(lst2[2])+7.5863825040)
    return lst2
```

```python
def ab_move(lst):
    lst3=lst[:]
    lst3[1]= str(float(lst3[1])+4.3800000000+8.76000000000)
    lst3[2]= str(float(lst3[2])+7.5863825040)
    return lst3
```

```python
with open('test.xyz','w') as f1:
    with open('Au-benzene-Au.xyz') as f:
        for k,v in enumerate(f):
            if k <=1:
                continue
            lst = v.strip().split()
            line1='   '.join(a_move(lst)) + '\n'
            line2='   '.join(b_move(lst)) + '\n'
            line3='   '.join(ab_move(lst)) + '\n'
            
            f1.write(v+line1+line2+line3)
#         a_move(f,f1)
#         b_move(f,f1)
#         ab_move(f,f1)
        f.close()
    f1.close()

with open('test.xyz') as f:
    print(f.read())
        
    
```

```python
group=['' for _ in range(4)]
with open('test.xyz','w') as f1:
    with open('2by2_S_(ch2)1_4tpa_C60_negf.xyz') as f:
        for k,v in enumerate(f):
            if k <=1:
                continue
            group[k%4] += v
        f1.write(''.join(group))
        f.close()
    f1.close()

with open('test.xyz') as f:
    print(f.read())
```
