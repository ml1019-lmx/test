```python
def str_float_z(num,height):
    return str(float(num)+height)
def str_float_x(num):
    return str(float(num)-9.5967844016 +9.14789)
def str_float_y(num):
    return str(float(num)-5.0535373223+5.02957 )

with open('2A_rightlead_molecule.xyz','w') as f1:
    with open('4_site-pos-1.xyz') as f:
        for k,v in enumerate(f):
            if k<=1:
                continue
            lst=v.strip().split()
            if k<=37:
                lst.append('L0')
            elif k <=99:
                lst.append('S')
            else:
                lst[2]=str_float_y(lst[2])
                lst[1]=str_float_x(lst[1])
                lst[3]=str_float_z(lst[3],32.36911-35.1193376278+2)
#                 lst.append('R0')
            line = '    '.join(lst) +'\n'
            f1.writelines(line)
        f.close()
    f1.close()
    
with open("test.xyz") as f:
    print(f.read())
```
