```python
import numpy as np
import math
def get_rms(records):
    """
    向量长度
    """
    return math.sqrt(sum([x ** 2 for x in records]))

```

```python
#input 2 atoms coordinate
B = np.array([0,0,0])
A = np.array([1,1,1])
```

```python
Zvector = A-B
Zvector
```

```python
# rotation matrix
Xvector = np.array([Zvector[0],Zvector[1],-(Zvector[0]**2+Zvector[1]**2)/Zvector[2]])
Yvector = np.array([Zvector[1],-Zvector[0],0])
P = np.array([Xvector/get_rms(Xvector),Yvector/get_rms(Yvector),Zvector/get_rms(Zvector)]).T
```

```python
P
```

```python
invP=np.linalg.inv(P)
invP
```

```python
with open("python_molecule_boat2_zdrection.xyz",'w') as f1:
    with open("S-ben-ch2-N-ch2^2-C60-pos-1.xyz") as f:
        for k,v in enumerate(f):
            if k == 0 or k == 1:
                line = v
            else:
                lst = v.strip().split()
                original = np.array([float(lst[1]),float(lst[2]),float(lst[3]) ]).reshape(3,1)
                new=np.dot(invP,original)
                for i in range(1,4):
                    lst[i] = str(new[i-1][0])
                line = '   '.join(lst)+"\n"
            f1.writelines(line)
        f.close()
    f1.close()
    
with open("python_molecule_boat2_zdrection.xyz") as f:
    print(f.read())
```
